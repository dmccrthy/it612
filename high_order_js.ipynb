```python
%%writefile code.js
// Part A
const data = [5, 12, 3, 20, 11, 9, 30];

// Use a for of loop over the numbers and check if >10 before adding
let result1 = [];
for (const number of data) {
  if (number > 10) {
    result1.push(number * 2);
  }
}

// here I use the filter to find >10 then map values to values*2
let result2 = data.filter((number) => number > 10).map((value) => value * 2);

console.log(`Loop: ${result1}`);
console.log(`Filter/Map: ${result2}`);

// Part B
const numbers = [1, 2, 3, 4, 5, 6];

// Due a similiar thing to part a just with the added reduce at the end.
let sum = numbers.filter((number) => number % 2 === 0).map((value) => value * 2).reduce((total, value) => total += value, 0);

console.log(`Sum: ${sum}`);

// Part C
const users = [
  { id: 1, name: "Alice", age: 25 },
  { id: 2, name: "Bob", age: 17 },
  { id: 3, name: "Carol", age: 19 }
];

console.log(users.some((user) => user.age < 18));
console.log(users.find((user) => user.id === 3));

users.forEach((user) => console.log(user));

// Part D
function repeat(n, action) {
  for (let i = 0; i < n; i++) {
    action(i);
  }
}

repeat(3, console.log); // 0, 1, 2
repeat(4, i => console.log("Step " + i));
```
